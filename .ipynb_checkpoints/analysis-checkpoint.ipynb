```python
import pandas as pd
import numpy as np
import math
import itertools
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# インプットデータを最低限整理
df = pd.read_csv("onlinenewspopularity.csv")
# df = data.drop(columns='url')
# 各カラム名の先頭にスペースがあったため、削除
tmp = []
for i in range(len(df.columns)): 
    tmp.append(df.columns[i].replace(" ", ""))
df.columns = tmp

df = df.drop(columns=['url'])

# 目的変数を対数変換
target = df['shares']
target_log = np.log(target)

# 対数変換したtargetに、列名をつける
target_log = pd.Series(target_log, name='shares_log')
# データフレームに追加
df = pd.concat([df, target_log], axis=1)
```

```python
# 外れ値を削除
q = df['shares'].quantile(0.95)

before_length = len(df)

df = df.query('shares < @q')

after_length = len(df)

# sns.boxplot(df['shares'])
# plt.show()
print('sharesが{}よりも大きい点を外れ値と判断し、削除した'.format(q))
print('外れ値削除前のデータ数は、{}で、削除後は{}です'
      .format(before_length, after_length))
```

# ダミー変数(曜日、ジャンル)を可視化

```python
# ダミー変数の効果を可視化する関数
def visual_dammy(df, ikichi, x, y): 
    pop = df[df[y] >= ikichi]
    unpop = df[df[y] < ikichi]
    pop_day = pop[x].sum().values
    unpop_day = unpop[x].sum().values
    
    # 図を作成（サイズを指定）
    fig = plt.figure(figsize=(26, 10))
    
    # タイトルを設定
    plt.title('Popular of Unpopular')
    
    # 棒グラフを作成
    plt.bar(np.arange(len(x)), pop_day, width=0.3, color='blue', label='Popular')
    # 2つの棒グラフを並べるため、0.3ずらす
    plt.bar(np.arange(len(x)) - 0.3, unpop_day, width=0.3, color='orange', label='Unpopular')
    
    plt.xticks(np.arange(len(x)), x)
    plt.xlabel('Dammy')
    plt.ylabel('Count')
    
    plt.legend()
    plt.tight_layout()
    plt.show()
    return unpop_day, pop_day
```

```python
# 曜日を可視化
x = df.columns.values[30:37]
m = df['shares'].median()
days = visual_dammy(df, m, x, 'shares')
days
```

->週末の方が人気である傾向にある

```python
# チャンネルを可視化
x = df.columns.values[12:18]
mean = df['shares'].mean()
channels = visual_dammy(df, mean, x, 'shares')
channels
```

# 調整済み散布図を見る（回帰分析するとき用）

```python
# 調整済み散布図を作成する
from sklearn import linear_model
def adjusted_variable_plot(xp_column, x_columns, y_column, data): 
    df = data
    # まず、xpを抜いた回帰モデルを作る
    X = df[x_columns]
    y = df[y_column]
    lr = linear_model.LinearRegression()
    lr.fit(X, y)
    # print(lr.score(X, y))

    # 残差を求める
    y_pred = lr.predict(X)
    y = y.values

    residuals_y = y_pred - y

    # 次に、yを抜いた変数にする回帰モデルを作る
    
    X2 = df[x_columns]
    y2 = df[xp_column]
    lr.fit(X2, y2)
    # print(lr.score(X2, y2))

    # 残差を求める
    y_pred2 = lr.predict(X2)
    y2 = y2.values

    residuals_x = y_pred2 - y2

    plt.scatter(residuals_x, residuals_y)
    plt.title('adjusted variable plot')
    plt.xlabel('residuals of ' + xp_column)
    plt.ylabel('residuals of shares')
    plt.grid()
    plt.show()
    
# shares = num_imgs + num_videosの回帰モデルに、
# num_hrefsを入れるかどうかを、この調整済み散布図を見て決める
adjusted_variable_plot('num_hrefs', ['num_imgs', 'num_videos'], 'shares', df)
```

```python
# 付属していたバイナリファイルの読み込み
def binary2str(): 
    # バイナリファイルの行数を取得
    def binary_size(): 
        cnt = 0
        with open('OnlineNewsPopularity.names', 'rb') as f: 
            for line in f: 
                cnt += 1
        return cnt
    
    size = binary_size()
    f = open("OnlineNewsPopularity.names", "rb")
    lines = f.readline()
    binary = []
    
    # 1行ずつ、バイナリを文字列に変換したものを読み込む
    for i in range(size): 
        binary.append(str(lines))
        lines = f.readline()
    f.close()
    return binary
a = binary2str()
```
