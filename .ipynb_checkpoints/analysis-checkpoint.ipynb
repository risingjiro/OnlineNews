```python
# ライブラリのimport
import pandas as pd
import numpy as np
import math
import itertools
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# インプットデータを最低限整理(カラム名の編集、目的変数の対数変換)
df = pd.read_csv("onlinenewspopularity.csv")
# df = data.drop(columns='url')
# 各カラム名の先頭にスペースがあったため、削除
tmp = []
for i in range(len(df.columns)): 
    tmp.append(df.columns[i].replace(" ", ""))
df.columns = tmp

df = df.drop(columns=['url'])

# 目的変数を対数変換
target = df['shares']
target_log = np.log(target)

# 対数変換したtargetに、列名をつける
target_log = pd.Series(target_log, name='shares_log')
# データフレームに追加
df = pd.concat([df, target_log], axis=1)
```

```python
# 外れ値を削除
q = df['shares'].quantile(0.95)

before_length = len(df)

df = df.query('shares < @q')

after_length = len(df)

# sns.boxplot(df['shares'])
# plt.show()
print('sharesが{}よりも大きい点を外れ値と判断し、削除した'.format(q))
print('外れ値削除前のデータ数は、{}で、削除後は{}です'
      .format(before_length, after_length))
```

# ダミー変数(曜日、ジャンル)を可視化

```python
# ダミー変数の効果を可視化する関数
def visual_dammy(df, ikichi, x, y): 
    pop = df[df[y] >= ikichi]
    unpop = df[df[y] < ikichi]
    pop_day = pop[x].sum().values
    unpop_day = unpop[x].sum().values
    
    # 図を作成（サイズを指定）
    fig = plt.figure(figsize=(26, 10))
    
    # タイトルを設定
    plt.title('Popular of Unpopular')
    
    # 棒グラフを作成
    plt.bar(np.arange(len(x)), pop_day, width=0.3, color='blue', label='Popular')
    # 2つの棒グラフを並べるため、0.3ずらす
    plt.bar(np.arange(len(x)) - 0.3, unpop_day, width=0.3, color='orange', label='Unpopular')
    
    plt.xticks(np.arange(len(x)), x)
    plt.xlabel('Dammy')
    plt.ylabel('Count')
    
    plt.legend()
    plt.tight_layout()
    plt.show()
    return unpop_day, pop_day
```

# 母比率の差の検定

https://bellcurve.jp/statistics/course/18227.html

帰無仮説は、1(平日)と2(休日)のシェアが多い割合は等しい
-> 平日と休日でシェアの数が等しくないことを示す。

両側検定を行う。有意水準を5%として、z_0.025は1.96。

```python
# sharesをダミー変数にする
median = df['shares'].median()

# 上位75%点
ikichi = 2400

df['shares_dammy'] = df['shares'].apply(lambda x: 0 if x < ikichi else 1)
```

```python
# 母比率の差の検定(有意水準を5%として、1.96(z_0.025)を超えていれば、帰無仮説は棄却)
def bohiritsu(n1, n2, p1, p2): 
    p_hat = ((n1 * p1) + (n2 * p2)) / (n1 + n2)

    upper = p1 - p2
    lowwer = np.sqrt((p_hat * (1 - p_hat)) * (1/n1 + 1/n2))
    # z = (p1 - p2) / np.sqrt(p_hat * (1 - p_hat) * (1/n1 + 1/n2))
    z = upper / lowwer
    return z
```

# 休日と平日で、シェアの多い割合を比較する

```python
# 休日のシェアのみ抽出
holiday = df['shares_dammy'][df['is_weekend'] == 1]
# 平日のシェアのみ抽出
weekday = df['shares_dammy'][df['is_weekend'] == 0]
```

```python
# 休日のシェアが高い行の割合
p_holiday = sum(holiday) / len(holiday)
n_holiday = len(holiday)
# 平日のシェアが高い行の割合
p_weekday = sum(weekday) / len(weekday)
n_weekday = len(weekday)

p1 = p_holiday
p2 = p_weekday
n1 = n_holiday
n2 = n_weekday

z = bohiritsu(n1, n2, p1, p2)
z
```

-> zが1.96よりも大きければ、帰無仮説は棄却(有意水準5%)

# t検定
平日のシェアの平均と、休日のシェアの平均に差があるかどうかを調べる
帰無仮説は、差がないこと

```python
# まず、平日とシェアの列を抽出
df_ordinary_days = df[["weekday_is_monday", "weekday_is_tuesday", 
                      "weekday_is_wednesday", "weekday_is_thursday", 
                     "weekday_is_friday", "shares"]]
# そして、どれかが1である行を抽出
df_ordinary_days = df_ordinary_days[(df_ordinary_days["weekday_is_monday"] == 1) | 
                (df_ordinary_days["weekday_is_tuesday"] == 1) | 
                (df_ordinary_days["weekday_is_wednesday"] == 1) | 
                (df_ordinary_days["weekday_is_thursday"] == 1) | 
                (df_ordinary_days["weekday_is_friday"] == 1)]

# 休日とシェアの列を抽出
df_holidays = df[["weekday_is_saturday", "weekday_is_sunday", "shares"]]
# どれかが1である行を抽出
df_holidays = df_holidays[(df_holidays["weekday_is_saturday"] == 1) | 
                (df_holidays["weekday_is_sunday"] == 1)]

"""
ordinary = df_ordinary_days['shares']
log_ordinary = np.log(ordinary)
plt.hist(log_ordinary)
plt.show()
"""

holidays = df_holidays['shares']
ordinary = df_holidays['shares']
```

```python
# シャピロウィルクによる正規性のチェック
import scipy.stats as stats
_, pa = stats.shapiro(log_ordinary)
_, pb = stats.shapiro(log_holidays)

print('平日のp値は{}, 休日のp値は{}'.format(pa, pb))

# 正規qq plot
stats.probplot(log_ordinary, dist='norm', plot=plt)
plt.show()

stats.probplot(log_holidays, dist='norm', plot=plt)
plt.show()
```

```python
def f_kentei(a, b): 
    a_var = np.var(a, ddof=1)
    b_var = np.var(b, ddof=1)
    
    a_jiyu = len(a) - 1
    b_jiyu = len(b) - 1
    
    f = a_var / b_var
    
    onesided_pval1 = stats.f.cdf(f, a_jiyu, b_jiyu)
    onesided_pval2 = stats.f.sf(f, a_jiyu, b_jiyu)
    twosided_pval = min(onesided_pval1, onesided_pval2) * 2
    
    print('F値: ', f)
    print('p値: ', twosided_pval)
    
def welch(a, b): 
    print(stats.ttest_ind(a, b, equal_var=False))
    
    
f_kentei(ordinary, holidays)
welch(ordinary, holidays)
```

```python
# 対応なしのステューデントのt検定
# 2つの標本の平均値に有意差がないことを帰無仮説とする
# これが棄却されること、つまり、有意差が出ることを期待する

def student_t(a, b): 
    t, p = stats.ttest_ind(a, b, equal_var=True)
    MU = abs(a.mean() - b.mean())
    SE = MU / t
    DF = len(a) + len(b) - 2
    CI = stats.t.interval(alpha=0.95, loc=MU, scale=SE, df=DF)
    
    print('対応なしt検定')
    print(f'p値 = {p:.3f}')
    print(f't値 = {t:.2f}')
    print(f'平均値の差   = {MU:.2f}')
    print(f'差の標準誤差 = {SE:.2f}')
    print(f'平均値の差の95%信頼区間CI = [{CI[0]:.2f} , {CI[1]:.2f}]')
    
student_t(holidays, ordinary)
```

# 調整済み散布図を見る（回帰分析するとき用）

```python
# 調整済み散布図を作成する
from sklearn import linear_model
def adjusted_variable_plot(xp_column, x_columns, y_column, data): 
    df = data
    # まず、xpを抜いた回帰モデルを作る
    X = df[x_columns]
    y = df[y_column]
    lr = linear_model.LinearRegression()
    lr.fit(X, y)
    # print(lr.score(X, y))

    # 残差を求める
    y_pred = lr.predict(X)
    y = y.values

    residuals_y = y_pred - y

    # 次に、yを抜いた変数にする回帰モデルを作る
    
    X2 = df[x_columns]
    y2 = df[xp_column]
    lr.fit(X2, y2)
    # print(lr.score(X2, y2))

    # 残差を求める
    y_pred2 = lr.predict(X2)
    y2 = y2.values

    residuals_x = y_pred2 - y2

    plt.scatter(residuals_x, residuals_y)
    plt.title('adjusted variable plot')
    plt.xlabel('residuals of ' + xp_column)
    plt.ylabel('residuals of shares')
    plt.grid()
    plt.show()
    
# shares = num_imgs + num_videosの回帰モデルに、
# num_hrefsを入れるかどうかを、この調整済み散布図を見て決める
adjusted_variable_plot('num_hrefs', ['num_imgs', 'num_videos'], 'shares', df)
```
